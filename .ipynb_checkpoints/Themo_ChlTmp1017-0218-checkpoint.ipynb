```python
import pandas as pd
import numpy as np # for log transformation
import matplotlib.pyplot as plt
from pandas.tseries import converter
```

```python
## The temperature dataset has a gap in data from '2017-10-01' to '2017-11-01'.
## This combined Chl/Temp will split both datasets into before and after segments 
## so as to better understand trends without blackout.
```

```python
## Import chlorophyll.
chl = pd.read_csv('/Users/regina/UH_projects/EM_TimeSeries/data/themo_chl0717-0618.csv', 
                 parse_dates=[[0,1]], index_col=0)
chl = chl.drop(columns=['turbidity_units', 'threshold', 'const_err'])
chl = chl[(chl.chlorophyll_concentration < 1.0) & (chl.chlorophyll_concentration > 0)]
chl = chl[:'2018-02-13'] # Remove bad data after February
chl = chl.drop(pd.Timestamp('2017-12-27 09:40:11')) 
chl_a = chl[:'2017-10-01'] # Data before gap
chl_a = chl_a[(chl_a.chlorophyll_concentration < 0.4)]
chl_b = chl['2017-11-01':] # Data after gap
chl_event = chl['2017-09-04':'2017-10-01'] # Data during Sept event
#chl_log = np.log(chl.chlorophyll_concentration) # for log transformation
```

```python
## Import temperature
tmp = pd.read_csv('/Users/regina/UH_projects/EM_TimeSeries/data/themo_sst0717-0618.csv',
                   parse_dates=[[0, 1]], index_col=0)
tmp = tmp.drop(columns=['s9_id', 'threshold', 'const_err'])
tmp = tmp[(tmp.temperature < 32) & (tmp.temperature > 20)]
tmp = tmp[:'2018-02-13'] # Data before battery blackout
tmp = tmp.drop(pd.Timestamp('2017-09-01 07:00:00')) 
tmp = tmp.drop(pd.Timestamp('2017-12-23 17:00:00')) 
tmp = tmp.drop(pd.Timestamp('2017-12-02 18:30:00')) # Individual outlier issues
tmp_a = tmp[:'2017-10-01'] # Data before gap
tmp_a = tmp_a[tmp_a.temperature > 25.5]
tmp_b = tmp['2017-11-01':] # Data after gap
tmp_event = tmp['2017-09-04':'2017-10-01'] # Data during Sept event
```

```python
# 
zone1a = tmp_a[tmp_a.depth < 30].rename(columns={"temperature": "zone1"})
zone1a = zone1a.drop(columns=['depth'])
zone1a_avg = zone1a.zone1.resample('D').mean()

zone1b = tmp_b[tmp_b.depth < 30].rename(columns={"temperature": "zone1"})
zone1b = zone1b.drop(columns=['depth'])
zone1b_avg = zone1b.zone1.resample('D').mean()

zone1event = tmp_event[tmp_event.depth < 30].rename(columns={"temperature": "zone1"})
zone1event = zone1event.drop(columns=['depth'])
zone1event_avg = zone1event.zone1.resample('D').mean()
```

```python
## Combined temperature and chlorophyll graphs

## Graph A, before battery blackout
combined_a = plt.figure(figsize=(14,10))
combined_a.suptitle('SST and Chlorophyll in the EM, August - October (A)',
                    fontsize=20)
chl_a_fig = combined_a.add_subplot(111)
chl_a_avg = chl_a.chlorophyll_concentration.resample('D').mean()
chl_a_fig.plot(chl_a, color='plum', alpha=0.4), chl_a_fig.plot(chl_a_avg, 
              color='purple', marker='o', markersize=5, linewidth=2)
chl_a_fig.set_ylabel('Chlorophyll Concentration', color='purple', fontsize=15)
chl_a_fig.set_xlabel('Date', fontsize=15)
chl_a_fig.legend(["Raw Chlorophyll Samples", "Chlorophyll Daily Average"], fontsize=13)
chl_a_fig.set_ylim([0,0.4])

tmp_a_fig = chl_a_fig.twinx()
tmp_a_avg = zone1a.resample('D').mean()
tmp_a_fig.plot(zone1a, color='turquoise', alpha= 0.4)
tmp_a_fig.plot(zone1a_avg, color='darkblue', linewidth=2, marker='o', markersize=5)
tmp_a_fig.set_ylabel('Temperature', color='darkblue', fontsize=15)
tmp_a_fig.legend(["Raw 30m Temperature Samples", "30m Temperature Daily Average"], fontsize=13)
tmp_a_fig.axvline('2017-09-07', color='tomato', linestyle='--')
tmp_a_fig.axvline('2017-09-18', color='tomato', linestyle='--')
```

```python
## Graph B, after battery blackout
combined_b = plt.figure(figsize=(14,10))
combined_b.suptitle('SST and Chlorophyll in the EM, November - February (B)', 
                    fontsize=20)
chl_b_fig = combined_b.add_subplot(111)
chl_b_avg = chl_b.chlorophyll_concentration.resample('D').mean()
chl_b_fig.plot(chl_b, color='plum', alpha=0.4), chl_b_fig.plot(chl_b_avg, 
              color='purple', marker='o', markersize=5, linewidth=2)
chl_b_fig.set_ylabel('Chlorophyll Concentration', color='purple', fontsize=15)
chl_b_fig.legend(["Raw Chlorophyll Samples", "Chlorophyll Daily Average"], fontsize=13, loc=2)
chl_b_fig.set_ylim([0,1.2])


tmp_b_fig = chl_b_fig.twinx()
tmp_b_avg = tmp_b.temperature.resample('D').mean()
tmp_b_fig.plot(zone1b, color='turquoise', alpha= 0.4)
tmp_b_fig.plot(zone1b_avg, color='darkblue', linewidth=2, marker='o', markersize=6)
tmp_b_fig.set_ylabel('Temperature', color='darkblue', fontsize=15)
tmp_b_fig.legend(["Raw 30m Temperature Samples", "30m Temperature Daily Average"], fontsize=13)
tmp_b_fig.set_ylim([20, 26])
```

```python
## Graph of event
event = plt.figure(figsize=(14,10))
event.suptitle('SST and Chlorophyll in the EM September 7-23', fontsize=20)
chl_event_fig = event.add_subplot(111)
chl_event_avg = chl_event.chlorophyll_concentration.resample('D').mean()
chl_event_fig.plot(chl_event, color='plum', alpha=0.4), 
chl_event_fig.plot(chl_event_avg, color='purple', marker='o', markersize=5, linewidth=2)
chl_event_fig.set_ylabel('Chlorophyll Concentration', color='purple', fontsize=15)
chl_event_fig.legend(["Raw Chlorophyll Samples", "Chlorophyll Daily Average"], fontsize=13, loc=2)
chl_event_fig.set_ylim([0,0.4])


tmp_event_fig = chl_event_fig.twinx()
#tmp_event_avg = tmp_event.temperature.resample('D').mean()
tmp_event_fig.plot(zone1event, color='turquoise', alpha= 0.4)
tmp_event_fig.plot(zone1event_avg, color='darkblue', linewidth=2, marker='o', markersize=5)
tmp_event_fig.set_ylabel('Temperature', color='darkblue', fontsize=15)
tmp_event_fig.legend(["Raw Temperature Samples", "Temperature Daily Average"], fontsize=13)
tmp_event_fig.set_ylim([28.4,31])
tmp_event_fig.axvline('2017-09-07', color='tomato', linestyle='--')
tmp_event_fig.axvline('2017-09-18', color='tomato', linestyle='--')
```
