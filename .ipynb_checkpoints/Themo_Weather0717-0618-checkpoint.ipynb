```python
import pandas as pd
from pandas.tseries import converter
import matplotlib.pyplot as plt
```

```python
## Import weather data, dropping unused columns. 
weather = pd.read_csv('/Users/regina/UH_projects/EM_TimeSeries/data/themo_weather0717-0618.csv', 
                 parse_dates=[[0,1]], index_col=0)
weather = weather.drop(columns=['threshold', 'const_err'])
weather = weather[:'2018-02-13']
weather_event = weather['2017-09-07':'2017-09-18']
weather_avg = weather.resample('D').mean()
```

```python
## Pressure
## There is a diurnal pressure variation!
## https://rmets.onlinelibrary.wiley.com/doi/pdf/10.1002/wea.857

pressure = plt.figure(figsize=(14,10))
pressure.suptitle('Pressure', fontsize=20)
pressure_fig = pressure.add_subplot(111)

pressure_fig.plot(weather.pressure, color='darkturquoise', alpha=0.4)
pressure_fig.plot(weather_avg.pressure, color='darkblue', linewidth=3,
                  marker='o', markersize=7)
pressure_fig.set_ylabel('Pressure in millibars', color='darkblue', fontsize=15)
pressure_fig.legend(['Raw Pressure Data', 'Average Daily Pressure Data'], fontsize=15,
                loc=2)

pressure_fig.axvline('2017-09-07', color='tomato', linestyle=':')
pressure_fig.axvline('2017-09-18', color='tomato', linestyle=':')

#pressure_hist = weather.pressure.hist(figsize=(13,8), bins=25, 
#                                      edgecolor='k')
#pressure_box = weather.boxplot('pressure', figsize=(13,7))
```

```python
## Air Temperatures
air = plt.figure(figsize=(14,10))
air.suptitle('Air Temperature', fontsize=20)
air_fig = air.add_subplot(111)

air_fig.plot(weather.temperature, color='blue', alpha=0.4)
air_fig.plot(weather_avg.temperature, color='darkblue', linewidth=3, 
             marker='o', markersize=7)
air_fig.set_ylabel('Air Surface Temperature in C', color='darkblue', fontsize=15)
air_fig.legend(['Raw Air Temperature Data', 'Air Temperature Daily Average'], fontsize=15, loc=0)
air_fig.set_ylim([8, 31])
air_fig.axvline('2017-09-07', color='tomato', linestyle='--')
air_fig.axvline('2017-09-18', color='tomato', linestyle='--')
```

```python
## Wind Speed
windspeed = plt.figure(figsize=(14,10))
windspeed.suptitle('Wind Speed', fontsize=20)
windspeed_fig = windspeed.add_subplot(111)

windspeed_fig.plot(weather.wind_speed, color='slategray', alpha=0.3)
windspeed_fig.plot(weather_avg.wind_speed, color='darkslategray', linewidth=3, 
             marker='o', markersize=7)
windspeed_fig.set_ylabel('Wind Velocity in m/s', color='midnightblue', fontsize=15)
windspeed_fig.legend(['Raw Wind Velocity Data', 'Wind Velocity Daily Average'], 
                     fontsize=15, loc=0)
windspeed_fig.set_ylim([-1, 24])
windspeed_fig.axvline('2017-09-07', color='tomato', linestyle='--')
windspeed_fig.axvline('2017-09-18', color='tomato', linestyle='--')
```

```python
## Humidity
humidity = plt.figure(figsize=(14,10))
humidity.suptitle('Air Humidity in Relative Humidity', fontsize=20)
humidity_fig = humidity.add_subplot(111)

humidity_fig.plot(weather.humidity, color='palegreen', alpha=0.5)
humidity_fig.plot(weather_avg.humidity, color='darkolivegreen', linewidth=3, 
             marker='o', markersize=7)
humidity_fig.set_ylabel('Humidity', color='darkolivegreen', fontsize=15)
humidity_fig.legend(['Raw Humidity Data', 'Humidity Daily Average'], 
                     fontsize=15, loc=0)
humidity_fig.axvline('2017-09-07', color='tomato', linestyle='--')
humidity_fig.axvline('2017-09-18', color='tomato', linestyle='--')
```

```python
## Wind Direction
direction = plt.figure(figsize=(14,10))
direction.suptitle('Air Direction in Degrees', fontsize=20)
direction_fig = direction.add_subplot(111)

direction_fig.plot(weather.humidity, color='magenta', alpha=0.3)
direction_fig.plot(weather_avg.humidity, color='indigo', linewidth=3, 
             marker='o', markersize=7)
direction_fig.set_ylabel('Direction', color='indigo', fontsize=15)
direction_fig.legend(['Raw Direction Data', 'Direction Daily Average'], 
                     fontsize=15, loc=0)
direction_fig.axvline('2017-09-07', color='tomato', linestyle='--')
direction_fig.axvline('2017-09-18', color='tomato', linestyle='--')
```
